```python
year=range(2000,2006)
```

```python
apples=[0.895,0.91,0.919,0.926,0.929,0.931]
oranges=[0.962,0.941,0.930,0.918,0.908,0.898]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield(tons per acre)')
plt.title('apple and orange yield / yield per year')
```

```python
apples=[0.895,0.91,0.919,0.926,0.929,0.931]
oranges=[0.962,0.941,0.930,0.918,0.908,0.898]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield(tons per acre)')
plt.title('apple and orange yield / yield per year')
plt.legend(['apples','oranges'])
```

```python
plt.plot(year,apples,marker='o')
plt.plot(year,oranges,marker='x')
plt.xlabel('year')
plt.ylabel('yield(tons per acre)')
plt.title('apple and orange yield / yield per year')
plt.legend(['apples','oranges'])
```

```python
plt.plot(year,apples,marker='s',c='b',linestyle='--',lw=2,ms=10,mew=2,mec='g')
plt.plot(year,oranges,marker='o',c='r',linestyle='-',lw=3,ms=10,mew=2,alpha=0.5)
plt.xlabel('year')
plt.ylabel('yield(tons per acre)')
plt.title('apple and orange yield / yield per year')
plt.legend(['apples','oranges'])
```

```python
plt.plot(year,oranges,'or')
plt.title('yield of oranges')
```

```python
plt.figure(figsize=(4,4))
plt.plot(year,apples,'or')
plt.title('yield of oranges')
```

```python
import matplotlib.pyplot as plt

x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title('bar chart')
plt.legend(['bar'])
plt.show()
```

```python
plt.scatter(x,y)
plt.title('scatter chart')
plt.legend(['scatter'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=[1,2,3,4]
e=(0.1,0.23,0.02,0.004)
plt.pie(x, explode= e)
plt.title('pie chart')
plt.show()
```
